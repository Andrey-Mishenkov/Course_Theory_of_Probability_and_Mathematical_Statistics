### Задача 3.

а). M(X)

где  
   $$a - M(X); {\sigma^2} - D(x); {\sigma} - std(X)$$

Наша формула в читаемом виде выглядит так:
$$f(x) = \frac{1}{4 \sqrt{2 \pi}} e ^ {- \frac{(x + a)^2}{32}}$$

тогда

### Задача 4.
Рост взрослого населения города X имеет нормальное распределение.  
Причем, средний рост равен 174 см, а среднее квадратичное отклонение равно 8 см.  
Какова вероятность того, что случайным образом выбранный взрослый человек имеет рост:
<small>

а) больше 182 см  
б) больше 190 см  
в) от 166 см до 190 см  
г) от 166 см до 182 см  
д) от 158 см до 190 см  
е) не выше 150 см или не ниже 190 см  
ё) не выше 150 см или не ниже 198 см  
ж) ниже 166 см.
</small>

Решаю с помощью правила 3 сигм:
<small>
* в промежутке от -1 сигма до +1 сигма - 68%    значений
* в промежутке от -2 сигма до +2 сигма - 95,4%  значений
* в промежутке от -3 сигма до +3 сигма - 99,72% значений
</small>

Получается, разброс значений роста:  
<small>
* от -1 сигма до +1 сигма - от 166 до 182 см
* от -2 сигма до +2 сигма - от 158 до 190 см
* от -3 сигма до +3 сигма - от 150 до 198 см
</small>

```python
p1 = 0.68
p2 = 0.954
p3 = 0.9972

print('Вероятность того, человек имеет рост:\n')
# а) больше 182 см
p = 0.5 - p1/2
print(f'\tа) больше 182 см;       \t\tp = {round(p, 5)};\tp = {1/2} - {p1/2}')

# б) больше 190 см
p = 1 - 0.5 - p2/2
print(f'\tб) больше 190 см;       \t\tp = {round(p, 5)};\tp = 1 - {1/2} - {p2/2}')

# в) от 166 см до 190 см
p = p1/2 + p2/2
print(f'\tв) от 166 см до 190 см; \t\tp = {round(p, 5)};\tp = {p1/2} + {p2/2}')

# г) от 166 см до 182 см
p = p1
print(f'\tг) от 166 см до 182 см; \t\tp = {round(p, 5)}')

# д) от 158 см до 190 см
p = p2
print(f'\tд) от 158 см до 190 см; \t\tp = {round(p, 5)}')

# е) не выше 150 см или не ниже 190 см
p = 1 - p3/2 - p2/2
print(f'\tе) не выше 150 см или не ниже 190 см; \tp = {round(p, 5)};\tp = 1 - {p3/2} - {p2/2}')

# ё) не выше 150 см или не ниже 198 см
p = 1 - p3
print(f'\tё) не выше 150 см или не ниже 198 см; \tp = {round(p, 5)};\tp = 1 - {p3}')

# ж) ниже 166 см
p = 0.5 - p1/2
print(f'\tж) ниже 166 см; \t\t\tp = {round(p, 5)};\tp = {1/2} - {p1/2}')
```

### Задача 5.
На сколько сигм (средних квадратичных отклонений) отклоняется рост человека, равный 190 см, от
математического ожидания роста в популяции, в которой M(X) = 178 см и D(X) = 25 кв.см?

```python
l = 190
M = 178
D = 25

sko = np.sqrt(D)
count_sko = (l - M) / sko

print(f'Рост человека {l} отклоняется от мат. ожидания {M} на {count_sko} сигм')
```
